```r
# Function to annotate columns with sample IDs and convert to a simple dataframe
process_zUMI_to_df <- function(zUMI_output, mapping_file_path) {
  # Load the processed metadata mapping file
  metadata_mapping <- read.csv(mapping_file_path)
  
  # Load Matrix package for handling sparse matrices
  if (!require("Matrix")) {
    install.packages("Matrix")
    library(Matrix)
  }
  
  # Create barcode to sample ID mapping
  barcode_to_sample <- setNames(as.character(metadata_mapping$sample_ID), 
                               as.character(metadata_mapping$barcode))
  
  # Extract the count matrix (using umicount exon all)
  count_matrix <- zUMI_output$umicount$exon$all
  
  # Rename columns using the mapping
  new_colnames <- colnames(count_matrix)
  for (barcode in names(barcode_to_sample)) {
    indices <- which(colnames(count_matrix) == barcode)
    if (length(indices) > 0) {
      new_colnames[indices] <- barcode_to_sample[barcode]
    }
  }
  colnames(count_matrix) <- new_colnames
  
  # Filter to keep only columns that start with "LC"
  lc_cols <- grep("^LC", colnames(count_matrix), value = TRUE)
  count_matrix_filtered <- count_matrix[, lc_cols, drop = FALSE]
  
  # Convert sparse matrix to regular matrix, then to dataframe
  counts_df <- as.data.frame(as.matrix(count_matrix_filtered))
  
  # Add gene IDs as a column
  counts_df$ENSEMBL <- rownames(counts_df)
  print(head(counts_df))
  # Return the filtered dataframe
  return(counts_df)
    
}

# Process the plate12 data
# First, load the original RDS file
plate12 <- readRDS("/ictstr01/home/icb/bhavishya.nelakuditi/bulk_seq/new_counts/zUMI_output_pto12/zUMIs_output/expression/bulk_seq_cgs_pto12.dgecounts.rds")

# Convert to dataframe and filter in one step
plate12_df <- process_zUMI_to_df(
  plate12, 
  "/ictstr01/home/icb/bhavishya.nelakuditi/bulk_seq/new_counts/cycle_cohort_metadata/outputs/plate_12_mapping_processed.csv"
)

# Show dimensions
cat("Dimensions of the filtered dataframe:", dim(plate12_df), "\n")

# Show the first few columns and rows
cat("First few columns and rows:\n")
print(head(plate12_df[, 1:min(5, ncol(plate12_df))]))  # Exclude gene_id column from preview
```

```r
library(biomaRt)
options(timeout = 300) 
# Set up the Ensembl connection
ensembl <- useEnsembl(biomart = "ensembl", dataset = "hsapiens_gene_ensembl")

# Get unique Ensembl IDs from your dataframe
ensembl_ids <- unique(plate12_df$ENSEMBL)

# Fetch gene symbols and descriptions from Ensembl
gene_annotations <- getBM(
  attributes = c("ensembl_gene_id", "external_gene_name", "description"),
  filters = "ensembl_gene_id",
  values = ensembl_ids,
  mart = ensembl
)

# Check how many IDs were successfully annotated
cat("Number of unique Ensembl IDs in dataset:", length(ensembl_ids), "\n")
cat("Number of Ensembl IDs with annotations:", nrow(gene_annotations), "\n")

# Merge annotations with the count data
plate12_df_annotated <- merge(
  plate12_df,
  gene_annotations,
  by.x = "ENSEMBL",
  by.y = "ensembl_gene_id",
  all.x = TRUE  # Keep all rows from plate12_df even if they don't have annotations
)

# Check the dimensions after merging
cat("Dimensions of the annotated dataframe:", dim(plate12_df_annotated), "\n")

# Show a sample of the annotated dataframe
cat("Sample of annotated dataframe:\n")
print(head(plate12_df_annotated[, c("ENSEMBL", "external_gene_name", "description", colnames(plate12_df_annotated)[1:3])]))
```

```r
col_names <- colnames(plate12_df_annotated)

# Find all columns that contain "ENSEMBL"
ensembl_cols <- grep("ENSEMBL", col_names)

# If there are duplicate ENSEMBL columns, keep only the first one
if (length(ensembl_cols) > 1) {
  plate12_df_annotated <- plate12_df_annotated[, -ensembl_cols[2:length(ensembl_cols)]]
  cat("Removed duplicate ENSEMBL columns.\n")
}

# Now filter out rows that don't have gene symbols
plate12_df_filtered <- plate12_df_annotated[!is.na(plate12_df_annotated$external_gene_name) & 
                                       plate12_df_annotated$external_gene_name != "", ]

# Check how many rows were removed
rows_before <- nrow(plate12_df_annotated)
rows_after <- nrow(plate12_df_filtered)
cat("Removed", rows_before - rows_after, "rows without gene symbols.\n")
cat("Dimensions after filtering:", dim(plate12_df_filtered), "\n")

# Show sample of the final dataframe
cat("Sample of final dataframe:\n")
print(head(plate12_df_filtered[, c("ENSEMBL", "external_gene_name", "description", 
                                 colnames(plate12_df_filtered)[4:6])]))
```
