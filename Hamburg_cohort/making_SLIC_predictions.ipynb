```r
library(caret)
library(glmnet)
library(ggplot2)
```

```r
uke_bulk_seq = read.csv("bulk_seq_uke_common_genes.csv")
```

```r
head(uke_bulk_seq)
```

```r
rownames(uke_bulk_seq) = uke_bulk_seq$sample_id
```

```r
uke_bulk_seq <- subset(uke_bulk_seq, select = -c(X,sample_id))
```

```r
head(uke_bulk_seq)
```

```r
rnaRes_top_500_genes = readRDS("top_500_genes_rnaRes.rds")
```

```r

best_model_index <- which.max(rnaRes_top_500_genes$r2Train)
best_model_index
# Extract the best model
best_model <- rnaRes_top_500_genes$modelList[[best_model_index]]
best_model
# Get the optimal lambda value for the best model
best_lambda <- rnaRes_top_500_genes$lambdaList[best_model_index]
best_lambda
```

```r
best_coefficients <- rnaRes_top_500_genes$coefMat[, best_model_index]
head(best_coefficients)
```

```r
# First, let's check how many genes we currently have
ncol(uke_bulk_seq) 

# Get the list of variables/genes that were used in training
# Assuming your model is a glmnet model
model_genes <- names(best_coefficients)  # This should give us all 23852 gene names

# Check which genes are missing in your current dataset
current_genes <- colnames(uke_bulk_seq)
missing_genes <- setdiff(current_genes, model_genes)
print(paste("Number of unwanted genes:", length(missing_genes)))

# Create a new dataframe with all required genes
# First, get the sample IDs (row names from your current data)
#sample_ids <- rownames(uke_bulk_seq)
uke_bulk_seqx = uke_bulk_seq[,model_genes]
ncol(uke_bulk_seqx)
```

```r
length(best_coefficients)
```

```r
complete_data = uke_bulk_seqx
```

```r
head(complete_data)
```

```r
write.csv(complete_data, "uke_top_500_genes_dataset_for_making_predictions.csv")
```

```r

best_model_index <- which.max(rnaRes_top_500_genes$r2Train)
best_model_index
# Extract the best model
best_model <- rnaRes_top_500_genes$modelList[[best_model_index]]
best_model
# Get the optimal lambda value for the best model
best_lambda <- rnaRes_top_500_genes$lambdaList[best_model_index]
best_lambda

set.seed(5863)
# Generate predictions on your data
predicted_values <- predict(best_model, newx = as.matrix(complete_data), s = best_lambda)
```

```r
as.data.frame(predicted_values)
```

```r
write.csv(as.data.frame(predicted_values), "predicted_values_top_500_factor3_genes.csv")
```
