```r
data <- read.csv("/ictstr01/home/icb/bhavishya.nelakuditi/mofa_workflow/results/03_results/03_Weight_Data_CGS_v3_MOFA.csv")

# Extract gene names after the double underscore
gene_names <- sapply(data$variable_name, function(x) {
  # Split by double underscore and take the second part
  parts <- strsplit(x, "__")[[1]]
  if(length(parts) > 1) {
    gene <- parts[2]
    # Remove version numbers (anything after a period)
    gene <- sub("\\.\\d+$", "", gene)
    return(gene)
  } else {
    return(NA)  # Return NA if no double underscore is found
  }
})

# Create a new data frame with the original variable names and the extracted gene names
result <- data.frame(
  original_name = data$variable_name,
  gene_name = gene_names,
  Factor3 = data$Factor3  
)


# Display the first few rows of the result
head(result)
```

```r
gene_df = result
```

```r

# First, filter out genes whose rownames start with "proteomic" or "clinical"
filtered_df <- gene_df[!grepl("^proteomic|^clinical", rownames(gene_df)), ]

# Sort by Factor3 in descending order to get highest positive values
top_positive <- filtered_df[order(filtered_df$Factor3, decreasing = TRUE), ]
# Take the top 250
top_positive_250 <- head(top_positive, 250)

# Sort by Factor3 in ascending order to get highest negative values
top_negative <- filtered_df[order(filtered_df$Factor3, decreasing = FALSE), ]
# Take the top 250
top_negative_250 <- head(top_negative, 250)

# Combine the two sets to get the top 500
top_500_genes <- rbind(top_positive_250, top_negative_250)
```

```r
top_500_genes
```

```r
top_500_factor3_genes = unique(top_500_genes$gene_name)
top_500_factor3_genes
```

```r
top_500_factor3_genes = intersect(top_500_factor3_genes, colnames(wide_cell_type_agnostic))
length(top_500_factor3_genes)
```
