```r
top_500_factor3_genes = read.csv("top_500_factor3_genes_matrix.csv")
```

```r
head(top_500_factor3_genes)
```

```r
rownames(top_500_factor3_genes) = top_500_factor3_genes$sample_id
top_500_factor3_genes <- subset(top_500_factor3_genes, select = -c(sample_id))
sampleOverlap  = intersect(rownames(top_500_factor3_genes), names(y))
sampleOverlap
length(sampleOverlap)
y <- y[sampleOverlap]
top_500_factor3_genes <- top_500_factor3_genes[sampleOverlap, ]
head(top_500_factor3_genes)
```

```r
library(caret)
# Now try the correlation again
cor_matrix <- cor(top_500_factor3_genes, method = "pearson", use = "complete.obs")
high_cor_cols <- findCorrelation(cor_matrix, cutoff = 0.9)
length(high_cor_cols) 
top_500_factor3_genes <- top_500_factor3_genes[, -high_cor_cols]
```

```r
head(top_500_factor3_genes)
```

```r
top_500_factor3_genes <- as.matrix(top_500_factor3_genes)
library(caret)
library(glmnet)
set.seed(5862)

rnaRes_top_500_factor3_genes <- runGlm(top_500_factor3_genes,y, repeats = 20, folds=5, method = "lasso", lambda ="lambda.1se", testRatio = 0.3)
```

```r
best_model_index <- which.max(rnaRes_top_500_factor3_genes$r2Train)
best_model_index
# Extract the best model
best_model <- rnaRes_top_500_factor3_genes$modelList[[best_model_index]]
best_model
# Get the optimal lambda value for the best model
best_lambda <- rnaRes_top_500_factor3_genes$lambdaList[best_model_index]
best_lambda
# Get the coefficients from the best model
best_coefficients <- rnaRes_top_500_factor3_genes$coefMat[, best_model_index]

# Create a data frame of non-zero coefficients (the selected genes)
selected_genes <- data.frame(
  gene_id = names(best_coefficients[best_coefficients != 0]),
  coefficient = best_coefficients[best_coefficients != 0]
) %>%
  arrange(desc(abs(coefficient)))

head(selected_genes)
```

```r
best_model_index <- which.max(rnaRes_top_500_factor3_genes$r2Train)
best_model <- rnaRes_top_500_factor3_genes$modelList[[best_model_index]]
best_lambda <- rnaRes_top_500_factor3_genes$lambdaList[best_model_index]
set.seed(5863)
# Generate predictions on your data
predicted_values <- predict(best_model, newx = as.matrix(top_500_factor3_genes), s = best_lambda)

# Create a data frame with actual and predicted values
prediction_results <- data.frame(
  Factor3 = y,
  predicted = predicted_values
)
prediction_results
```

```r
colnames(prediction_results)[colnames(prediction_results) == "s1"] <- "proxy_top_500_factor3_genes"
```
